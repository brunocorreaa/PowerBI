```python
""" Geração de dados fictícios da tabela de fEstoque """

# Importação
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Configurações do Dataset
n_linhas = 100000
data_inicial = datetime(2026, 1, 1)

# Gerando dados fictícios
data = {
    'Filial': np.random.randint(1, 101, size=n_linhas), # Filiais de 1 a 100
    'Produto': np.random.randint(100000, 100101, size=n_linhas), # Produtos de 1000000 a 1000100
    'Data': [(data_inicial + timedelta(days=np.random.randint(0, 60))).strftime('%d/%m/%Y') for _ in range(n_linhas)],
    'Qtd_Estoque': np.random.randint(10, 5000, size=n_linhas),
    'CML_Estoque': np.round(np.random.uniform(500, 80000, size=n_linhas), 2)
}

# Criando o DataFrame e salvando em Excel
df_estoque = pd.DataFrame(data)
df_estoque.to_excel('fEstoque.xlsx', index=False)

print("Arquivo 'fEstoque.xlsx' gerado")
```

```python
""" Geração de dados ficticios de vendas"""

# Importação
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Configurações do Dataset
n_vendas = 200000
data_inicial = datetime(2026, 1, 1)

# Gerando dados consistentes
data = {
    # Corrigido: Todos agora usam n_vendas (200.000)
    'Filial': np.random.randint(1, 101, size=n_vendas), 
    'Produto': np.random.randint(100000, 100101, size=n_vendas), 
    'Data': [(data_inicial + timedelta(days=np.random.randint(0, 60))).strftime('%d/%m/%Y') for _ in range(n_vendas)],
    'Qtd_Venda': np.random.randint(1, 15, size=n_vendas),
}

# Criando DataFrame
df_vendas = pd.DataFrame(data)

# Gerando valores financeiros
# Venda_CML: Custo Médio Líquido
df_vendas['Venda_CML'] = np.round(np.random.uniform(50, 500, size=n_vendas), 2)
# Venda_Valor_Liquido: Valor de venda (aplicando margem sobre o CML)
df_vendas['Venda_Valor_Liquido'] = np.round(df_vendas['Venda_CML'] * np.random.uniform(1.3, 1.8, size=n_vendas), 2)

# Salvando em CSV
# Para 200k linhas, CSV é muito mais rápido que Excel para o Power BI processar
df_vendas.to_csv('fVendas.csv', index=False, sep=';', encoding='latin1')

print("Finalizado! Arquivo 'fVendas.csv'")
```

```python
""" Geração de dados ficticios de produtos"""

# Importação
import pandas as pd
import numpy as np

# Configuração: IDs de 100.000 até 100.100 (Total 101 itens)
ids_produtos = np.arange(100000, 100101)
n_produtos = len(ids_produtos)

# Definição da Hierarquia Solicitada
setores = ['Setor_A', 'Setor_B', 'Setor_C', 'Setor_D']
grupos_map = {
    'Setor_A': ['Grupo_AA', 'Grupo_AB'],
    'Setor_B': ['Grupo_BA', 'Grupo_BB'],
    'Setor_C': ['Grupo_CA', 'Grupo_CB'],
    'Setor_D': ['Grupo_DA', 'Grupo_DB']
}

data_produtos = []

for p_id in ids_produtos:
    # Seleção lógica baseada na hierarquia
    setor = np.random.choice(setores)
    grupo = np.random.choice(grupos_map[setor])
    
    # Gerando nomes de Classe e Subclasse seguindo o padrão AAA, AAAA
    prefixo = grupo.split('_')[1] # Pega 'AA', 'BA', etc.
    classe = f"Classe_{prefixo}A" if p_id % 2 == 0 else f"Classe_{prefixo}B"
    subclasse = f"Subclasse_{prefixo}AA" if p_id % 2 == 0 else f"Subclasse_{prefixo}AB"
    
    # Nome do produto ex: 100000 - Produto_AA_100000
    nome_produto = f"Produto_{prefixo}_{p_id}"
    
    data_produtos.append({
        'Produto': p_id,
        'Nome_Produto': nome_produto,
        'Subclasse': subclasse,
        'Classe': classe,
        'Grupo': grupo,
        'Setor': setor
    })

# Criando o DataFrame
df_dprodutos = pd.DataFrame(data_produtos)

# Exportando para Excel
df_dprodutos.to_excel('dProdutos_Final.xlsx', index=False)

print(f"Tabela dProdutos gerada")
```
